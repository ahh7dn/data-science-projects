# Function to define a list of questions
Takes an xml document as argument. Returns a list of questions.

```python
def dataframe_of_q_and_a(xml_filename):
    doc = xml.dom.minidom.parse(xml_filename)
    ilvl_list = doc.getElementsByTagName('w:ilvl')
    highlighted_text = doc.getElementsByTagName('w:highlight')
    all_text = doc.getElementsByTagName("w:t")
    
    # select all except the last question and compile into list
    questions = []
    for element in ilvl_list:
        if element.getAttribute('w:val') == "0":
            questions.append(element)
    question_list = []
    for element in questions:
        question_cousin = element.parentNode
        question_uncle = question_cousin.parentNode
        question_parent = question_uncle.nextSibling
        if question_parent.lastChild != None:
            question = question_parent.lastChild
            question_list.append(question)
    
    if len(question_list) == 99:
        # Add last question to list
        last_question = questions[-1].parentNode.parentNode.nextSibling.nextSibling.nextSibling.lastChild
        question_list.append(last_question)
    
    q_a_list = []
    answer_list = []
    q_a_dict = dict({})
    # i iterates trhough index values for each element in all_text
    for i in np.arange(len(all_text)):
        current_index = i
        current_text = all_text[current_index]
        next_index = i+1
        #if next_index in np.arange(len(all_text)):
        #    next_text = all_text[next_index]
        # Save question as dictionary key
        if current_text in question_list:
            question = current_text.toxml()
        if current_text not in question_list:
            if current_text.previousSibling.lastChild in highlighted_text:
                answer_list.append(current_text.toxml())
        if next_index in np.arange(len(all_text)):
            next_text = all_text[next_index]
            if next_text in question_list:
                q_a_dict[question] = answer_list
                answer_list = []
        if next_index not in np.arange(len(all_text)):
            if current_text.previousSibling.lastChild in highlighted_text:
                answer_list.append(current_text.toxml())
            q_a_dict[question] = answer_list
            
    q_a_dict_correct = dict({})
    for i in question_list:
        q_a_dict_correct[i.toxml()] = q_a_dict[i.toxml()]
    
    if len(q_a_dict_correct) == 100:
        q_a_series = pd.Series(q_a_dict_correct, name=xml_filename)
        q_a_dataframe = q_a_series.to_frame().T
    else:
        print('Not all questions are represented')
        
    return(q_a_dataframe)
```
